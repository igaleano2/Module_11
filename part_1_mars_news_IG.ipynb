```python
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd

browser=Browser('chrome')
```

```python
# Visit the URL
url = "https://static.bc-edx.com/data/web/mars_news/index.html"
browser.visit(url)

# Parse the HTML
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
##print(soup)
```

```python
# List to contain article info
articles_list = []
```

```python
# Extract all the text elements
titles = soup.find_all('div','content_title')
previews = soup.find_all('div','article_teaser_body')
print(titles)
```

```python
head_articles=[]
for title in titles:
    head= title.get_text()
    data1={'Title':head}
    #print(title)
    head_articles.append(data1)
head_articles    
```

```python
body_articles=[]
for preview in previews:
    body= preview.get_text()
    data2={'preview':body}
    #print(title)
    body_articles.append(data2)
body_articles    
```

```python
body_articles
```

```python
nasa_articles= list(zip(head_articles,body_articles))
nasa_articles

df_nasa_articles=pd.DataFrame(nasa_articles,columns=['Title','Article'])

df_nasa_articles
```
